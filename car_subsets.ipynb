```python
def all_makes(df):
    parent_dict = {}
    for makers in df.make.unique():
        parent_dict[makers] = {}
        
    return parent_dict

def all_models(df, parent_dict):
    for make, model_dict in parent_dict.items():
        for model in df[df.make == make].model.unique():
            model_dict[model] = {}
            
    return parent_dict

def all_years(df, parent_dict):
    for make, model_dict in parent_dict.items():
        for model, year_dict in model_dict.items():
            for year in df[(df.make ==make) & (df.model==model)].year.unique():
                year_dict[year] = {}
                
    return parent_dict

def all_prices(df, parent_dict):
    for make, model_dict in parent_dict.items():
        for model, year_dict in model_dict.items():
            for year, price_dict in year_dict.items():
                price_dict['avg_price'] = df[(df.make == make) & (df.model==model) & (df.year == year)].price.mean()
                price_dict['min_price'] = df[(df.make == make) & (df.model==model) & (df.year == year)].price.min()
                price_dict['max_price'] = df[(df.make == make) & (df.model==model) & (df.year == year)].price.max()
                price_dict['num_vehicles'] = df[(df.make==make)&(df.model==model)&(df.year==year)].price.size

    return parent_dict            
    
more_than_5_vehicles = []

    
parent = all_makes(df)
parent = all_models(df, parent)
parent = all_years(df, parent)
parent = all_prices(df, parent)


for make, model_dict in parent.items():
    for model, year_dict in model_dict.items():
        for year, price_dict in year_dict.items():
            if price_dict['num_vehicles'] > 5:
                more_than_5_vehicles.append({'make': make, 'model': model, 'year': year})

more_than_5_vehicles
```

```python
def mk_mod_yr_df(item):
    return df[(df.make == item['make'])& (df.model == item['model'])&(df.year == item['year'])]
items = more_than_5_vehicles
mk_mod_yr_df(items[0])
```
